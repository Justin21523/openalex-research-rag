# OpenAlex Research Intelligence — EDA
Exploratory analysis of the sample works corpus.

```
import duckdb
import polars as pl
import plotly.express as px

conn = duckdb.connect("data/openalex.duckdb")
```

```
conn.execute("SELECT COUNT(*) as works FROM works").df()
```

```
df = conn.execute("SELECT publication_year, COUNT(*) as count FROM works GROUP BY 1 ORDER BY 1").df()
px.bar(df, x="publication_year", y="count", title="Works per Year")
```

```
conn.execute("SELECT work_id, title, cited_by_count FROM works ORDER BY cited_by_count DESC LIMIT 10").df()
```
