# Modeling — BM25, Embeddings, Hybrid Search, RAG

```
import os; os.environ["HF_HOME"] = os.path.expanduser("~/.cache/huggingface")
from src.utils.db import get_connection
from src.features.bm25_index import BM25Index
from src.retrieval.vector_store import VectorStore
from src.features.embeddings import WorkEmbedder
from src.retrieval.hybrid_search import HybridSearch
from src.models.rag_pipeline import RAGPipeline
```

```
conn = get_connection()
bm25 = BM25Index().load()
vs = VectorStore()
emb = WorkEmbedder()
search = HybridSearch(bm25, vs, emb)
```

```
results = search.search("transformer attention mechanism", k=5, mode="hybrid")
for r in results: print(r["rank"], r["work_id"], f"rrf={r[chr(39)]rrf_score"]:.4f}")
```

```
rag = RAGPipeline(search, conn)
result = rag.answer_extractive("What are graph neural networks?")
print(result.answer_text[:400])
print("Citations:", result.citations)
```
